```python
def plot_surface(clf, X, y):
    plot_step = 0.01
    palette = sns.color_palette(n_colors=len(np.unique(y)))
    cmap = ListedColormap(palette)
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))
    plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=cmap, alpha=0.3)

    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap, alpha=.7,
                edgecolors=np.array(palette)[y], linewidths=2)
```

```python
from sklearn.metrics import accuracy_score
plt.figure(figsize=(15, 4))
for i, (x, y) in enumerate(datasets):
    plt.subplot(1, 3, i + 1)
    X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=.3, random_state=0)
    dt = DecisionTreeClassifier(random_state=0).fit(X_train, y_train)
    y_test_predict = dt.predict(X_train)
    y_train_predict = dt.predict(X_test)
    print(f'Accuracy score for X_train of {i+1} dataset = {accuracy_score(y_test_predict, y_train)}')
    print(f'Accuracy score for X_train of {i+1} dataset = {accuracy_score(y_train_predict, y_test)}', '\n')
    plot_surface(dt, x, y)
```

Реализуйте функцию find_best_split из модуля hw5code.py

```python
df = pd.read_csv('students.csv')
df.head()
```

```python
def entropy(left, right, sub_total):
    if left == 0 or right == 0:
        return 0
    answer = -right/sub_total*np.log2(right/sub_total) - left/sub_total*np.log2(left/sub_total)
    return answer
```

```python
def find_best_split(feature, target):
    test = sorted(feature.unique())
    test = [test[0]-1] + test + [test[0]+1]
    bins = []

    full = feature.shape[0]
    pos = target.sum()
    neg = full - pos
    s0 = entropy(neg, pos, full)

    x = []
    y = []

    for i in range(0, len(test)-1):
        bins.append((test[i]+test[i+1])/2)


    for b in bins:
        tmp = target[feature>b]

        right_total = tmp.shape[0]
        left_total = full - right_total
        right_pos = tmp.sum()
        right_neg = right_total - right_pos
        left_pos = pos - right_pos
        left_neg = neg - right_neg

        s_left = entropy(left_neg, left_pos, left_total)
        s_right = entropy(right_neg, right_pos, right_total)
        IG = s0 - left_total/full * s_left - right_total/full * s_right
        x.append(b)
        y.append(IG)
        
    position = np.nanargmax(y)
    best_IG = y[position]
    best_split = x[position]
    
    return x, y, best_split, best_IG
    
```

```python
plt.figure(figsize=(12,9))
for i in range(1, 5):
    feature = df.iloc[:, i]
    target = df.iloc[:, 5]
    thresh, gini, thresh_best,gini_best = find_best_split(feature, target)
    plt.plot(thresh, gini, label=df.columns[i] )
    plt.legend()
    plt.xlabel('thresholds')
    plt.ylabel('entropy')
```
